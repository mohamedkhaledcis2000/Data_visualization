```python
import pandas as pd
import numpy as np
import math
```

```python
div = pd.read_csv('div.csv')
# return a new column with 1 or 0 values with condition Total Docks are greater than 10 or less than it
div['classificated']=np.where(div['Total Docks']>10,1,0)
div.head(10)
```

```python
div['Docks in Service'] = div['Docks in Service'] + 100 
div['Docks in Service'].head(10)
```

```python
div['Docks in Service'] = div['Docks in Service']*1.5
```

```python
div['Docks in Service'].head(10)
```

```python
# rounded values ofcolumn
div['Docks in Service']= div['Docks in Service'].apply(math.ceil)
```

```python
# rounded values of column and put it in newo column
div['Docks rounded'] =div['Docks in Service'].apply(math.ceil)
```
